## 1. Merge sort

```python
def merge(left, right):
    '''
    Given two sorted list, return the merged sorted list.
    '''
    ''' YOUR CODE HERE '''
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    while i < len(left):
        result.append(left[i])
        i += 1
    while j < len(right):
        result.append(right[j])
        j += 1
    return result

assert merge([], []) == []
assert merge([], [1, 2, 3]) == [1, 2, 3]
assert merge([1, 3, 5], [2, 4, 6]) == [1, 2, 3, 4, 5, 6]
```

```python
#BT1:
def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0

    # Duyệt qua cả hai danh sách và thêm phần tử nhỏ hơn vào merged_list
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # Nếu còn phần tử trong list1, thêm chúng vào merged_list
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # Nếu còn phần tử trong list2, thêm chúng vào merged_list
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

# Ví dụ sử dụng:
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]

print(merge_sorted_lists(list1, list2))
```

```python
def merge_sort(nums):
    '''
    sort a list using merge sort algorithm
    hint: use merge method you have written before.
    '''
    ''' YOUR CODE HERE '''
    if len(nums) < 2:
        return nums[:]
    else:
        middle = len(nums) // 2
        left = merge_sort(nums[:middle])
        right = merge_sort(nums[middle:])
        return merge(left, right)

assert merge_sort([]) == []
assert merge_sort([1]) == [1]
assert merge_sort([6, 3, 2, 4, 5, 1]) == [1, 2, 3, 4, 5, 6]
```

```python
#BT2:
def merge_sort(arr):
    # Nếu danh sách chỉ có một phần tử hoặc rỗng, trả về danh sách đó
    if len(arr) <= 1:
        return arr

    # Tìm điểm giữa của danh sách
    mid = len(arr) // 2

    # Đệ quy chia danh sách thành hai nửa và sắp xếp chúng
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    # Hợp nhất hai nửa đã sắp xếp
    return merge_sorted_lists(left_half, right_half)

# Ví dụ sử dụng:
unsorted_list = [38, 27, 43, 3, 9, 82, 10]
sorted_list = merge_sort(unsorted_list)
print(sorted_list)
```

## 2. Newton's method
Problem: Find $x$ such that $f(x) = 0$.
Algorithm:  
  Initialize $x_0$  
  Repeat the process until a sufficiently value is reached:  
        $x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)}$

### 2.1. Find square root

```python
def sqrt(y, epsilon=1e-15):
    '''Write function to compute y'''
    assert y > 0
    x = 1
    '''YOUR CODE HERE'''
    while abs(x**2 - y) >= y * epsilon:
        x = 0.5 * (x + y / x)
    return x
```

```python
print (sqrt(1))
print (sqrt(2))
print (sqrt(25))
print (sqrt(100))
```

```python
#BT3
def can_bac_2_newton(x, epsilon=1e-15):
    if x < 0:
        return "Không có nghiệm thực"
    else:
        guess = x / 2  # Đoán ban đầu
        while abs(guess ** 2 - x) > epsilon:
            guess = (guess + x / guess) / 2
        return guess

print (can_bac_2_newton(100))
print (can_bac_2_newton(25))
print (can_bac_2_newton(2))
```

### 2.2 Find one root of the equation $4x^4+5x^3-2x^2+3x+7=1000$

```python
def f(x):
    ''' Evaluate f(x)'''
    '''YOUR CODE HERE'''
    return 4 * x**4 + 5 * x**3 - 2 * x**2 + 3 * x + 7 - 1000
```

```python
def df(x):
    ''' Evaluate f'(x)'''
    '''YOUR CODE HERE'''
    return 16 * x**3 + 15 * x**2 - 4 * x + 3
```

```python
def find_root(epsilon=1e-10):
    '''Find one root of the given equation.'''
    '''YOUR CODE HERE'''
    x = -10000.0
    while abs(f(x)) >= epsilon:
        x = x - f(x) / df(x)
    return x
x = find_root()
print ("x =  ", x)
print ("f(x) = ", f(x))
```

```python
#BT4
def f(x):
    return 4*x**4 + 5*x**3 - 2*x**2 + 3*x + 7 - 1000

def f_prime(x):
    return 16*x**3 + 15*x**2 - 4*x + 3

def newton_method(f, f_prime, initial_guess, epsilon=1e-10, max_iterations=100):
    x = initial_guess
    iteration = 0
    while True:
        iteration += 1
        x_next = x - f(x) / f_prime(x)
        if abs(x_next - x) < epsilon or iteration >= max_iterations:
            break
        x = x_next
    return x

initial_guess = 0  # Đoán ban đầu
root = newton_method(f, f_prime, initial_guess)
print("Một nghiệm của phương trình là:", root)
```
